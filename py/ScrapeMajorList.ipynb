```python
from bs4 import BeautifulSoup
from urllib import request
import json
```

```python
html = request.urlopen('https://admissions.illinois.edu/Discover/Academics/majors_alpha').read()
soup = BeautifulSoup(html, 'html.parser')
link_lists = soup.find_all('ul', 'link-list')
list_of_majors = []
for i in range(len(link_lists)):
    majors = link_lists[i].findAll('a')
    for j in range(len(majors)):
        print(majors[j].text)
        list_of_majors.append(majors[j].text)
```

```python
with open('../json/majors.json', 'w') as jsonfile:
    json.dump(list_of_majors, jsonfile)
```
